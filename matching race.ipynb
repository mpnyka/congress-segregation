```python
staffer_names = pd.read_csv("columbia/core_person2.csv") 
```

```python
staffer_names[["first_name", "middle_name"]] = staffer_names["first_name"].str.split(expand=True, n=1)
```

```python
staffer_names
```

```python
race_staffer2010 = pred_fl_reg_name(staffer_names, 'last_name', 'first_name')
```

```python
race_staffer2010.to_csv("race_staffer.csv")
```

```python
race_staffer2010_df = pd.DataFrame(race_staffer2010)
adf = race_staffer2010_df.groupby(['race']).agg({'last_name': 'count'})
adf.unstack().apply(lambda r: r / r.sum(), axis=1).style.format("{:.2%}")
```

```python
adf.apply(lambda r: r / r.sum())
```

```python
race_staffer_wiki = pred_wiki_name(staffer_names, 'last_name', 'first_name')
```

```python
race_staffer_wiki
```

```python
race_staffer_wiki.to_csv("race_staffer_wiki.csv")
```

```python
race_staffer_wiki_df = pd.DataFrame(race_staffer_wiki)
adf = race_staffer_wiki_df.groupby(['race']).agg({'last_name': 'count'})
adf.apply(lambda r: r / r.sum())
```
